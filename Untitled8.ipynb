```python
import re
import os
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from tensorflow.keras.layers import Dense, Input, Dropout
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import LSTM, Dense, Dropout, RepeatVector, BatchNormalization
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.regularizers import l2

# Paths for the Data
Attack_free_dataset_Path = "Attack_free_dataset.txt"
DoS_attack_dataset_Path = "DoS_attack_dataset.txt"
Fuzzy_attack_dataset_Path = "Fuzzy_attack_dataset.txt"
Impersonation_attack_dataset_Path = "Impersonation_attack_dataset.txt"

SplitRatio = 0.8

# Function to extract features from files
def extract_features(file_name):
    ids, data_values = [], []

    with open(file_name, 'r') as file:
        for line in file:
            line = line.strip()

            # Extract ID
            id_match = re.search(r'ID:\s*([0-9A-Fa-f]{4})', line)
            if id_match:
                ids.append(int(id_match.group(1), 16))  # Convert hex ID to decimal
            else:
                continue  # Skip if ID not found

            # Extract Data
            data_match = re.search(r'DLC:\s*\d+\s*(.*)', line)
            if data_match:
                hex_values = data_match.group(1).strip().split()
                valid_hex_values = []
                # Validate and convert hex values to integers
                for value in hex_values:
                    try:
                        valid_hex_values.append(int(value, 16))  # Convert hex to integer
                    except ValueError:
                        continue  # Skip invalid values
                data_values.append(valid_hex_values)

    # Ensure both lists are of the same length
    min_len = min(len(ids), len(data_values))
    return np.array(ids[:min_len]), data_values[:min_len]  # Trim to the shorter length

# Extract features from each dataset
AttackFree_features = extract_features(Attack_free_dataset_Path)
DoSAttack_features = extract_features(DoS_attack_dataset_Path)
FuzzyAttack_features = extract_features(Fuzzy_attack_dataset_Path)
ImpersonationAttack_features = extract_features(Impersonation_attack_dataset_Path)

# Combine features and labels for training
combined_features = []
combined_labels = []

for label, features in zip([0, 1, 2, 3],
                           [AttackFree_features, DoSAttack_features, FuzzyAttack_features, ImpersonationAttack_features]):
    ids, data_values = features
    for i in range(len(ids)):
        # Append features with variable-length data_values
        combined_features.append([ids[i]] + data_values[i])  # Keep it flexible
        combined_labels.append(label)

# Convert combined labels to numpy arrays
combined_labels = np.array(combined_labels)

# Pad data_values to ensure consistent input shape
max_length = max(len(item) for item in combined_features)  # Find the max length for padding
padded_features = pad_sequences(combined_features, maxlen=max_length, padding='post')

# Convert to numpy arrays
padded_features = np.array(padded_features)

# Normalize the features
scaler = StandardScaler()
padded_features = scaler.fit_transform(padded_features)

# Shuffle the combined features and labels
padded_features, combined_labels = shuffle(padded_features, combined_labels, random_state=42)

# Split the data into train and test sets
def split_data(data, Ratio):
    num_rows = data.shape[0]
    num_train_rows = int(np.floor(num_rows * Ratio))
    return data[:num_train_rows], data[num_train_rows:]

# Split the data
train_features, test_features = split_data(padded_features, SplitRatio)
train_labels, test_labels = split_data(combined_labels, SplitRatio)

print(f'Train Features Shape: {train_features.shape}')
print(f'Test Features Shape: {test_features.shape}')

# Convert labels to one-hot encoding
if train_labels.size > 0:
    train_labels = to_categorical(train_labels)

if test_labels.size > 0:
    test_labels = to_categorical(test_labels)

# Custom Focal Loss function
def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)  # Avoid division by zero
        y_true = tf.cast(y_true, tf.float32)
        cross_entropy = -y_true * tf.math.log(y_pred)
        loss = alpha * tf.math.pow(1. - y_pred, gamma) * cross_entropy
        return tf.reduce_mean(tf.reduce_sum(loss, axis=-1))
    return focal_loss_fixed

# Define the Neural Network architecture
def ANN(input_shape):
    model = Sequential()
    model.add(Input(shape=(input_shape,)))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(4, activation='softmax'))  # 4 output classes
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5)
    model.compile(optimizer=optimizer, loss=focal_loss(gamma=2., alpha=0.75), metrics=['accuracy'])
    return model

# Build the model
input_shape = train_features.shape[1]
ann_model = ANN(input_shape)

# Print model summary
print("** Model Summary **")
ann_model.summary()

# Early Stopping to avoid overfitting
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5)

# Adjust weight for class 3 (Impersonation)
class_weight = {0: 1.0, 1: 3.0, 2: 3.0, 3: 2.0}

# Train the model
history = ann_model.fit(train_features, train_labels, epochs=1,
                        callbacks=[early_stopping],
                        batch_size=32, validation_split=0.2, verbose=1, class_weight=class_weight)

# Evaluate the model on the test data
test_loss, test_accuracy = ann_model.evaluate(test_features, test_labels)

print(f'Test Loss: {test_loss}')
print('Test Accuracy: {:.2f} %'.format(np.round(test_accuracy, 3) * 100))

# Get ANN predictions
ann_predictions = ann_model.predict(test_features)
predicted_classes = np.argmax(ann_predictions, axis=1)
true_classes = np.argmax(test_labels, axis=1)

# Get classification report
print("\nClassification Report for ANN:")
print(classification_report(true_classes, predicted_classes))

# Confusion Matrix for ANN
conf_matrix = confusion_matrix(true_classes, predicted_classes)

# Normalize the confusion matrix by rows (true labels)
conf_matrix_normalized = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

# Plot normalized confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_normalized, annot=True, fmt=".2f", cmap="Blues", xticklabels=['Attack Free', 'DoS', 'Fuzzy', 'Impersonation'],
            yticklabels=['Attack Free', 'DoS', 'Fuzzy', 'Impersonation'])
plt.title('Normalized Confusion Matrix for ANN')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Get ANN predictions
ann_predictions = ann_model.predict(test_features)
predicted_classes = np.argmax(ann_predictions, axis=1)  # Get the predicted class for each sample

# Store Attack-Free predicted labels
attack_free_predictions = test_features[predicted_classes == 0]  # Class '0' is Attack Free

# Print the number of Attack-Free predictions
print(f'Number of samples predicted as Attack Free: {attack_free_predictions.shape[0]}')

# You can now use the attack_free_predictions variable for further analysis or processing.
plt.title('Normalized Confusion Matrix for ANN')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

'''
# Define the LSTM model
def LSTM_Model(input_shape):
    model = Sequential()
    model.add(LSTM(64, input_shape=(input_shape, 1), return_sequences=True))
    model.add(LSTM(32))
    model.add(Dense(4, activation='softmax'))

    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Reshape for LSTM
train_features_lstm = np.expand_dims(train_features, axis=2)
test_features_lstm = np.expand_dims(test_features, axis=2)

# Build and train the LSTM model
lstm_model = LSTM_Model(train_features_lstm.shape[1])
history_lstm = lstm_model.fit(train_features_lstm, train_labels, epochs=1,
                              batch_size=32, validation_split=0.2, verbose=1)

# Get LSTM predictions on the test set
lstm_predictions = lstm_model.predict(test_features_lstm)
lstm_predicted_classes = np.argmax(lstm_predictions, axis=1)
true_classes = np.argmax(test_labels, axis=1)

# Evaluate LSTM performance
print("\nLSTM Classification Report:")
print(classification_report(true_classes, lstm_predicted_classes))

# Find incorrect LSTM predictions
incorrect_predictions = np.where(lstm_predicted_classes != true_classes)[0]

# Use incorrect predictions to augment ANN training
feedback_features = test_features[incorrect_predictions]
feedback_labels = test_labels[incorrect_predictions]

# Retrain ANN with LSTM feedback
lstm_features = np.expand_dims(lstm_predictions, axis=1)
augmented_train_features = np.hstack((train_features, lstm_features))
ann_model.fit(augmented_train_features, train_labels, epochs=1,
              callbacks=[early_stopping], batch_size=32, verbose=1, class_weight=class_weight)

# Confusion Matrix for LSTM
lstm_conf_matrix = confusion_matrix(true_classes, lstm_predicted_classes)

# Normalize the LSTM confusion matrix by rows (true labels)
lstm_conf_matrix_normalized = lstm_conf_matrix.astype('float') / lstm_conf_matrix.sum(axis=1)[:, np.newaxis]

# Plot normalized confusion matrix for LSTM
plt.figure(figsize=(8, 6))
sns.heatmap(lstm_conf_matrix_normalized, annot=True, fmt=".2f", cmap="Greens",
            xticklabels=['Attack Free', 'DoS', 'Fuzzy', 'Impersonation'],
            yticklabels=['Attack Free', 'DoS', 'Fuzzy', 'Impersonation'])

plt.title('Normalized Confusion Matrix for LSTM')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
'''




# Set TensorFlow environment variables
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_CPP_MIN_LOG_LEVEL']  = '2'

# Path for the Data
train_file_path = 'Attack_free_dataset.txt'
test_file_path = attack_free_predictions

# Hyperparameters
time_steps = 1
features = 9
batch_size = 64
epochs = 1
dropout_rate = 0.15  # Slightly reduced dropout rate to retain more information
l2_lambda = 0.001   # L2 regularization factor

# Function to extract features from a single file
def extract_features(file_name):
    ids, data_values = [], []

    with open(file_name, 'r') as file:
        for line in file:
            # Return a copy of the string with leading and trailing whitespace removed
            line = line.strip()

            # Extract ID
            id_match = re.search(r'ID:\s*([0-9A-Fa-f]{4})', line)
            if id_match:
                ids.append(int(id_match.group(1), 16))  # Convert hex ID to decimal

            # Extract Data
            data_match = re.search(r'DLC:\s*\d+\s*(.*)', line)
            if data_match:
                # Convert hex values to integers
                hex_values = data_match.group(1).strip().split()
                valid_hex_values = []

                # Validate and convert hex values to integers
                for value in hex_values:
                    try:
                        valid_hex_values.append(int(value, 16))  # Convert hex to integer
                    except ValueError:
                        continue  # Skip invalid values

                data_values.append(valid_hex_values)

    return np.array(ids), data_values  # Keep data_values as a list of lists

# Function to preprocess data (padding, scaling, reshaping)
def preprocess_data(ids, data_values, time_steps, features):
    # Combine features
    combined_features = []
    for i in range(len(ids)):
        # Append features with variable-length data_values
        combined_features.append([ids[i]] + data_values[i])  # Keep it flexible

    # Pad data_values to ensure consistent input shape
    max_length = max(len(item) for item in combined_features)  # Find the max length for padding
    padded_features = pad_sequences(combined_features, maxlen=max_length, padding='post')

    # Convert to numpy arrays
    padded_features = np.array(padded_features)

    # Normalize the features
    scaler = StandardScaler()
    padded_features = scaler.fit_transform(padded_features)

    # Reshape the data to match LSTM expected input shape (batch_size, time_steps, features)
    reshaped_features = padded_features.reshape((padded_features.shape[0], time_steps, features))

    return reshaped_features, scaler

# Extract features for training data
train_ids, train_data_values = extract_features(train_file_path)
train_features, scaler = preprocess_data(train_ids, train_data_values, time_steps, features)

# Shuffle the training data
train_features = shuffle(train_features, random_state=42)

# Create the LSTM-autoencoder model
model = Sequential([
    LSTM(256, return_sequences=True, input_shape=(time_steps, features), kernel_regularizer=l2(l2_lambda)),
    BatchNormalization(),
    Dropout(dropout_rate),

    LSTM(128, return_sequences=True, kernel_regularizer=l2(l2_lambda)),
    BatchNormalization(),
    Dropout(dropout_rate),

    LSTM(64, kernel_regularizer=l2(l2_lambda)),
    Dropout(dropout_rate),

    RepeatVector(time_steps),

    LSTM(64, return_sequences=True, kernel_regularizer=l2(l2_lambda)),
    Dropout(dropout_rate),

    LSTM(128, return_sequences=True, kernel_regularizer=l2(l2_lambda)),
    BatchNormalization(),
    Dropout(dropout_rate),

    LSTM(256, return_sequences=True, kernel_regularizer=l2(l2_lambda)),
    Dropout(dropout_rate),

    Dense(features, activation='linear')  # Use 'linear' for continuous outputs instead of sigmoid
])

# Compile the model with gradient clipping, learning rate scheduler, and MAE as an additional metric
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001, clipnorm=1.0)  # Gradient clipping to prevent exploding gradients
model.compile(optimizer=optimizer, loss='mse', metrics=[tf.keras.metrics.MeanAbsoluteError()])

# Learning rate scheduler callback
lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)

# Early stopping callback
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history = model.fit(train_features, train_features,
                    batch_size=batch_size,
                    epochs=epochs,
                    validation_split=0.2,
                    callbacks=[early_stopping, lr_scheduler])

# Extract features for test data
test_features, _ = preprocess_data(np.zeros(attack_free_predictions.shape[0]), attack_free_predictions, time_steps, features)

# Evaluate the model on the test data
test_loss, test_mae = model.evaluate(test_features, test_features, batch_size=batch_size)

print(f"Test Loss: {test_loss}")
print(f"Test Mean Absolute Error (Accuracy Proxy): {test_mae}")
```
